## Visualization of charging sites

```python
# Specify the file path and name
file_path = ".mapbox_token"

# Open the file in write mode
# If the file does not exist, it will be created
with open(file_path, "w") as file:
    file.write("sk.eyJ1Ijoia2ltZG9oYW5pY2kiLCJhIjoiY2x1eHJxdWUxMHJiajJsbWdoeGc5ZHhhZyJ9.kVDodSW47e9HM6Rw8bf0iw")

# Set up the token
mapbox_token = "sk.eyJ1Ijoia2ltZG9oYW5pY2kiLCJhIjoiY2x1eHJxdWUxMHJiajJsbWdoeGc5ZHhhZyJ9.kVDodSW47e9HM6Rw8bf0iw"

# Specify the token in the code
px.set_mapbox_access_token(mapbox_token)

# Read the Excel file into a pandas DataFrame
site_data = pd.read_excel("/Users/kimdohanici/Documents/04 Master/6. Masterthesis/Optimization Model/Data IONITY/Station Data & Utilization/site_data_DE_new.xlsx")

# Create the scatter map plot using Plotly Express
fig = px.scatter_mapbox(site_data, 
                        lat="latitude",
                        lon="longitude",
                        color="price_per_kwh",  # Color by price per kWh
                        size="utilization",  # Size by utilization
                        zoom=4,
                        color_continuous_scale=px.colors.sequential.Oranges, 
                        hover_name="name", 
                        title="IONITY Charging Stations in Germany",  
                        animation_frame="month", 
                        hover_data={"demand": ":.2f",  
                                    "cost_variable": ":.2f",  
                                    "site_profit": ":.2f", 
                                    },
                        labels={"latitude": "Latitude", "longitude": "Longitude", "month": "Month", "utilization": "Utilization", 
                                "price_per_kwh": "Price per kWh", "demand": "Charging Demand", "cost_variable": "Operational Cost",
                                "demand": "Adhoc Charging Demand", "site_profit": "Profit",
                                },  
                        )

# Customize the plot
fig.update_layout(mapbox_style="open-street-map",  
                coloraxis_colorbar=dict(title="Price per kWh"),  
                legend_title="Charging Points",  
                margin={"l": 20, "r": 20, "t": 70, "b": 10},
                width=1100,  
                height=650
                )

# Show the plot
fig.show()
```
